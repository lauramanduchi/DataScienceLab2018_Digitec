# Product distribution

```python
import matplotlib.pyplot as plt
```

Distribution of product per category:

```python
pCat = pd.read_sql_query('''
          select "ProductTypeId", count(distinct "ProductId") from product_only_ids
          group by "ProductTypeId"
          ''', c)
```

```python
pCat
```

```python
#select which category you want to plot
index_min = 1
index_max = 20
```

```python
pCat.iloc[index_min:index_max].plot.bar(x="ProductTypeId", y="count", rot=0, figsize =(30,10))
plt.show()
```

Number of products bought for all category:

```python
pTotalDist = pd.read_sql_query(''' 
select "Items_ProductId", sum("Items_ItemCount") from product_purchase 
group by "Items_ProductId";
''', c)
```

```python
pTotalDist
```

```python
pTotalDist.iloc[index_min:index_max].plot.bar(x="Items_ProductId", y="sum", figsize =(20,10))
plt.show()
```

Given a category, the distribution of products bought:

```python
pt = pd.read_sql_query('''
          select distinct "ProductId" from product_only_ids
          where "ProductTypeId" = '6'
          order by "ProductId"
          ''', c)
```

```python
pCatDist = pTotalDist.loc[pTotalDist['Items_ProductId'].isin(pt["ProductId"])]
pCatDist
```

```python
pCatDist.iloc[index_min:index_max].plot.bar(x="Items_ProductId", y="sum", figsize =(20,10))
plt.show()
```

Note: out of 1445 products just 483 has been bought!

Given a category, the distribution of products per filters (questions / PropertyType):

```python
pQ = pd.read_sql_query('''
          select "PropertyTypeId", count(distinct "ProductId") from product_only_ids
          where "ProductTypeId" = '1099'
          group by "PropertyTypeId"
          ''', c)
pQ
```

Given a category, the distribution of products per values (question + answers / PropertyType + PropertyValueId):

```python
pQA = pd.read_sql_query('''
          select "PropertyTypeId", "PropertyValueId" count(distinct "ProductId") from product
          where "ProductTypeId" = '1099'
          group by "PropertyTypeId", "PropertyValueId"
          ''', c)
pQA
```
